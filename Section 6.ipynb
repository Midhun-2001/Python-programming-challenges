## Program 63 - Dress sizes (part 2)

```python
def midhun(new,stock):
   
    while True:
        print("Do you wish to add (A) or remove (R) a dress?")
        dress=input()
        if dress=='A':
            new=int(input("Please enter a new dress size: "))
       
            if new==999:
                break
            else:
                stock.append(new)
                print("Stock =",stock)
                continue
            
       
        elif dress=='R' :
            new=int(input("Please enter a new dress size: "))
            stock.remove(new)
            if new==999:
                break
            else:
                print("Stock =",stock)
                continue
            
        
    print("Stock =",stock)
    largest=max(stock)
    smallest=min(stock)
    print("The largest dress in the stock is size",largest)
    print("The smallest dress in the stock is size",smallest)
    
    
    
stock=[]
new=0    
midhun(new,stock)
```

## Program 64 - Charity Collection (part 5)

```python
def charity(first):
   
    raised=0  
    print("Enter the total raised by each:")
    for i in range(first):
        total=int(input())
        raised=raised+total
    

    if raised<1000:
        double=100+raised
        print("A total of $"+str(raised),"was raised.")
        print("With the company bonus,this is $"+str(double)+".")
    elif raised>=1000 and raised<=2000:
        double=2*raised
        print("A total of $"+str(raised),"was raised.")
        print("With the company bonus,this is $"+str(double)+".")
    elif raised>2000:
        double=4000+(raised-2000)
        print("A total of $"+str(raised),"was raised.")
        print("With the company bonus,this is $"+str(double)+".")

print("How many charity raisers were there?")
   
first=int(input())
charity(first)
```

## Program 65 - Cycling Speed (part 5)

```python
print("What is the circumfarance of your wheel in millimeters?")
c=float(input())
print("How many journeys do you wish to enter")
j=int(input())
wheels=[]
minutes=[]
for i in range(j):
    print("How many wheel revolutions have taken place in your journey",i+1)
    w=float(input())
    wheels.append(w)
    print("How many minutes did you cycle for journey",i+1)
    m=float(input())
    minutes.append(m)
W=sum(wheels)
M=sum(minutes)

def total_dis(c,W):
    d=(c*W)/(10**6)
    return d
k=total_dis(c,W)
def av_speed(k,M):
    s=(k)/(M/60)
    return s
    
print("You covered",total_dis(c,W),"km.")
print("At an average speed of",av_speed(k,M),"kmh.")

    
```
